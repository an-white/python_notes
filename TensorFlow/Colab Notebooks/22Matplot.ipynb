```
import matplotlib.pyplot as plt, numpy as np

# la libreria plt permite imprimir en pantalla los vectores y matrices
# generados en python

# el modulo .plot permite generar un objeto el cual contiene una grafica de lo que este entre ()
fx=plt.plot([0,1,4,11,26])

# .show()muestra lo contenido entre () puede ser, vector, matriz o funcion
plt.show(fx)

x=np.linspace(0,10,20)
y=x**2

# dentro () se puede definir el color, espesor y forma de graficar el vector que se desea graficar
# tambien si se def 2 variables dentro del plot una se tomara como el eje x y la otra como el eje y, y se 
# graficara ese vector
# al definir otros valores se puede utilizar abreviaciones para cambiar color, añadir marcadores y cambiar diseño de linea
# esto se puede definir entre entre "oo-" siendo el primer valor la inicial de color
# el 2do el timpo de marcador o para circular s para cuadrado etc, y el ultimo para el tipo de linea

plt.plot(x,y, "oo-", linewidth=2, markesize=4)
```
