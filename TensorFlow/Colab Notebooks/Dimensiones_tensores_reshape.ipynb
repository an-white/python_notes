**Dimensiones, tensores y reshape**

**Tipos de dimensiones**

**Escalar**
Cuando es un unico dato dimension 0

**Vector**
Cuando es una lista, o tuplas de datos dimension 1

**Matriz**
Un DatFrames de 2 dimensiones Ejemplos con caracteristicas o datapoints contra caracteristicas

**Tensor**
Tiene 3 a n dimensiones

Tensores de serie de tiempo 
(cantidad de ejemplos, Caracteristicas de los ejemplos, Caracteristicas de los ejemplos respecto al tiempo)

Tensor de 4D:
Para Imagenes en RGB
Cantidad de ejemplos, alto, ancho, Canales RGB

```python
import numpy as np
```

```python
scalar=np.array(23)
print(scalar.ndim)
print(scalar.shape)
```

```python
vector=np.array([22,23])
print(vector.ndim)
print(vector.shape)
```

```python
matriz=np.array([[22,23,22],[11,33,44]])
print(matriz.ndim)
print(matriz.shape)
```

```python
tensor=np.array([
                 [[22,23,22],[11,33,44]],
                 [[22,23,22],[11,33,44]],
                 [[22,23,22],[11,33,44]],
                 ])
print(tensor.ndim)
print(tensor.shape)
```

```python
#reshape
x=np.array([[0,1],
            [2,3]])
print(x.shape)
print(x)

print(f"\n{x.reshape(4,1)}")
print(f"\n{x.reshape(1,4)}")

print(f"\n{x.transpose()}")
```
