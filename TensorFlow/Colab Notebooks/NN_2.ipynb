**Funciones de activacion con numpy**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def step(x):
  return np.piecewise(x,[x<0.0,x>0.0],[0,1]) # funcion escalon
```

```python
x = np.linspace(10,-10,100)
```

```python
plt.plot(x,sigmoid(x))
```

```python
plt.plot(x,step(x))
```

```python
def relu(x):
  return np.piecewise(x,[x<0.0,x>0.0],[0,lambda x: x]) # buscar por que se usa la funcion lambda para poder def valores de x
```

```python
plt.plot(x,relu(x))
```

```python
def tanh(x):
  return (np.exp(x)- np.exp(-x))/(np.exp(-x)+np.exp(x))
```

```python
plt.plot(x,tanh(x))
```

Funcion de Coste o perdida

```python
def MSE(y,y_hat, derivate=False):
  if derivate:
    return y_hat - y
  else:
    return np.mean((y_hat-y)**2)
```

```python
vector = np.array([0,0,0,1,1])
prediction = np.array([0.1,0.5,0.2,0.0,1.1])
```

```python
MSE(vector,prediction)
```
